```python
import pandas as pd
import re
from collections import Counter
from itertools import chain

df = pd.read_csv("trackercommondata_2024_2025_only_educational.csv")

s = df["Trackers"].fillna("").astype(str).str.strip()
lists_ = s.apply(lambda x: [t for t in re.split(r"[;,|\s]+", x) if t])
with open("tracker_lists_repr.txt", "w", encoding="utf-8") as f:
    for lst in lists_:
        f.write(repr(lst) + "\n")
# Flatten all trackers into one list
all_trackers = list(chain.from_iterable(lists_))

# Count frequency of each tracker
freq = Counter(all_trackers)

# Save to txt: one tracker per line as "tracker\tcount"
with open("tracker_frequency.txt", "w", encoding="utf-8") as f:
    for tracker, count in sorted(freq.items(), key=lambda x: (-x[1], x[0])):
        f.write(f"{tracker}\t{count}\n")
counts = lists_.apply(len)


uniq = set(t.strip() for lst in lists_ for t in lst)

total_trackers = int(counts.sum())
unique_trackers = len(uniq)
mean_val = counts.mean()
std_val = counts.std(ddof=1)
median_val = float(counts.median())
max_val = int(counts.max())
zero_pct = float((counts == 0).mean() * 100)


summary = pd.DataFrame([{
    "#Tracker (total)": total_trackers,
    "Unique Trackers": unique_trackers,
    "Mean ± SD": f"{mean_val:.2f} ± {std_val:.2f}",
    "Median": median_val,
    "Max Tracker": max_val,
    "Zero Tracker %": round(zero_pct, 2),
}])
print("trackercommondata_2025_only_educational.csv")
print(summary)

summary.to_csv("racker_stats_single_file.csv", index=False)
counts.to_frame("tracker_count").to_csv("tracker_counts_single_file.csv", index=False)

```

```python
import pandas as pd

# Load the datasets for 2024 and 2025
data_2024 = pd.read_csv("2024.csv")
data_2025 = pd.read_csv("2025.csv")

# Clean the tracker columns by filling NaN values with empty strings
data_2024['Trackers'] = data_2024['Trackers'].fillna('')
data_2025['Trackers'] = data_2025['Trackers'].fillna('')

# Identify websites with no trackers in 2024 and with trackers in 2025
no_tracker_2024 = data_2024[data_2024['Trackers'] != '']
tracker_2025 = data_2025[data_2025['Trackers'] == '']
print(no_tracker_2024['Website'])
print("-----")
print(tracker_2025['Website'])
# Merge the datasets to get websites that have trackers in 2025 but not in 2024
websites_with_tracker_in_2025 = pd.merge(tracker_2025, no_tracker_2024, on='Website', how='inner')

# Save the result to a CSV file
websites_with_tracker_in_2025[['Website', 'Trackers_x', 'Trackers_y']].to_csv('websites_with_tracker_in_2025.csv', index=False)
websites_with_tracker_in_2025['Website'].to_csv('websites_list.csv',index=False)
print("The list has been saved to 'websites_with_tracker_in_2025.csv'.")
```
